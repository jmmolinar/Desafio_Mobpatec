PREGUNTA 10: En algunos pedidos, no está registrada la “Fecha de envío”, por tanto se deben completar esos registros como la “Fecha compra” + mediana de la diferencia entre “Fecha de envio” - “Fecha compra” del resto de pedidos.

```python
#NECESITO UN DATAFRAME SOLO CON LAS FECHA DE ENVÍO NULAS que resultan ser 6

df_ventas_solo_fecha_envio_null = df_ventas[df_ventas["Fecha envío"].isnull()]
df_ventas_solo_fecha_envio_null
```

```python
#NECESITO UN DATAFRAMA SIN LAS FECHAS ENVIO NULAS para poder hacer con el RESTO DE LOS PEDIDOS:
# 1. Fecha envío - Fecha compra
# 2. Mediana(Fecha envío - Fecha compra)
# 3. Luego asignar a los NULOS de FECHA ENVÍO el resultado de:
#    SU Fecha Compra + Mediana(Fecha envío - Fecha compra) DEL DATAFRAME DONDE NO HAY FECHAS ENVÍO NULO

# el df_ventas tiene 9961 registros, por lo tanto, al quitarle los 6 nulos de Fecha envío
# Nuestro data frame sin nulos de Fecha envío tendrá 9955

df_ventas_sin_fecha_envio_null = df_ventas[df_ventas["Fecha envío"].notnull()]
df_ventas_sin_fecha_envio_null.info()
```

```python
# 1. Fecha envío - Fecha compra

df_ventas_sin_fecha_envio_null["Dif. Fechas"] = df_ventas_sin_fecha_envio_null.apply(lambda x: x["Fecha envío"]-x["Fecha compra"], axis=1)
df_ventas_sin_fecha_envio_null.head()
```

```python
# 2. Mediana(Fecha envío - Fecha compra)

mediana_restantes = df_ventas_sin_fecha_envio_null["Dif. Fechas"].median()
mediana_restantes
```

```python
# 3. Luego asignar a los NULOS de FECHA ENVÍO el resultado de:
#    SU Fecha Compra + Mediana(Fecha envío - Fecha compra) DEL DATAFRAME DONDE NO HAY FECHAS ENVÍO NULO

# Para ello Primero tomo el data frame que saqué solo con los nulos en Fecha Envío para sumarle su FECHA COMPRA
# CON LA MEDIANA DE DIFERENCIAS DE FECHAS PERO DEL DATA FRAME QUE CONTIENE LOS RESTANTES DONDE NO HAY FECHA ENVÍO NULL

df_ventas_solo_fecha_envio_null["Fecha envío"] = df_ventas_solo_fecha_envio_null["Fecha compra"] + mediana_restantes
df_ventas_solo_fecha_envio_null
```

```python
# Se puede observar en ésta celda anterior que a las Fechas Envío que eran nulas se les asignó el valor de
# su Fecha Compra + los 4 días que resultaron de la Mediana(Fecha envío - Fecha compra) de los restantes que si tenían
# todas sus Fechas Envío
```

```python
# AHORA DEBO CONCATENAR en:

# df_ventas_act a df_ventas_sin_fecha_envio_null junto con df_ventas_solo_fecha_envio_null

df_ventas_act = pd.concat([df_ventas_solo_fecha_envio_null, df_ventas_sin_fecha_envio_null], ignore_index=False)
df_ventas_act.head(80)
```

```python
#PLUS - agregando la Diferencia de Fechas para que aparezcan en los registros donde antes la Fecha Envío era Nula

df_ventas_act['Dif. Fechas'].fillna(mediana_restantes, inplace=True)
df_ventas_act.head(80)
```

```python
#PLUS - Ordenando el DATA FRAME DE VENTAS POR EL INDICE

df_ventas_act.sort_index()

# Tenemos de nuevo el Data Frame de Ventas Actualizado con sus 9961 registros 
# y todas las Fechas de envío completas además de las Dif. Fechas
```

PREGUNTA 13: Teniendo en cuenta el precio de venta unitario, las unidades vendidas y el coste unitario, nos solicitan obtener el beneficio obtenido en cada uno de los pedidos. Para ello debe crear una nueva columna “Beneficio”.

¿Cuál es la categoría de productos que supone un mayor beneficio?

```python
df_global.head()
```

```python
#TOTAL VENTA = PRECIO UNITARIO * CANTIDAD
#BENEFICIO = TOTAL VENTA - (COSTO PRODUCCIÓN * CANTIDAD)

df_global["Beneficio"] = df_global.apply(lambda x: x["Total venta"]-(x["Coste Producción"]*x["Cantidad"]), axis=1)
df_global.head()
```

```python
#PLUS - Viendo un resumen con la suma de beneficios por Categoría para cada Pedido

df_global.pivot_table(index="ID Pedido", columns="Categoría", values="Beneficio", aggfunc="sum")
```

PREGUNTA 17: La empresa se está planteando abrir nuevas plantas de producción en aquéllas ubicaciones donde la diferencia entre “Fecha envío” y “Fecha compra” es más elevada para reducir los tiempos. ¿En qué países elegiría construir estas nuevas plantas de producción?

```python
#En df_global_rpc tengo todos los datos que necesito para responder ésta pregunta

df_global_rpc.info()
```

```python
#DataFrame para ver diferencias de Fechas entre compra y envío por paises

df_plantas_paises = pd.DataFrame(df_global_rpc.loc[:,["País","Fecha compra","Fecha envío","Dif. Fechas"]])
df_plantas_paises["Días"] = df_plantas_paises["Dif. Fechas"].dt.days
df_plantas_paises
```

```python
#Arriba tuve que separar los días de la diferencia de fechas en una columna aparte
#Para poder utilizar los días como enteros y determinar la media

df_plantas_paises.info()
```

```python
#Observo la media de días de diferencia entre compra y envío por cada país

df_plantas_paises.groupby("País")["Días"].agg("mean")
```

```python
#El criterio que voy a utilizar para crear nuevas plantas, es hacerlo en aquellos países 
#donde su promedio de diferencia de dias sea
#mayor al promedio de diferencia de días de todas las ventas

df_plantas_paises["Días"].mean()
```

```python
df_pais_dias = pd.DataFrame(df_plantas_paises.groupby("País")["Días"].agg("mean"))
df_pais_dias > df_plantas_paises["Días"].mean()
```

```python
df_pais_dias[df_pais_dias> df_plantas_paises["Días"].mean()]

#Con estas dos últimas sentencias se puede ya observar los países con diferencias de días mayores a la media
```

```python
#Sin embargo, creemos un DataFrame más ordenado para mostrar los resultados de una forma más elegante

array_pais_planta = np.array(df_pais_dias.index)
array_pais_planta

array_dias_planta = np.array(df_pais_dias["Días"])
array_dias_planta

dic_pais_dias = {"País": array_pais_planta,
            "Diferencia días": array_dias_planta}

df_pais_dias_dif_mean = pd.DataFrame(dic_pais_dias)
df_pais_dias_dif_mean
```

```python
def nuevaPlanta(x):
    if x > df_plantas_paises["Días"].mean():
        z = "Si"
    else:
        z = "No"
    return z

df_nuevas_plantas = df_pais_dias_dif_mean
df_nuevas_plantas["¿Crear plantas nuevas?"] = df_nuevas_plantas["Diferencia días"].apply(nuevaPlanta)
df_nuevas_plantas
```

PREGUNTA 18: Como resultado final, el CEO de su compañía necesita disponer de un reporte en formato Excel con el beneficio de cada categoría en cada uno de los países. Exporta el resultado a un fichero con nombre “Reporte Beneficios” y pestaña “Beneficios_cat_pais”.

```python
df_global_rpc.head()
```

```python
#Haré un primer formato con PIVOT

df_reporte_beneficios_formato_1 = df_global_rpc.pivot_table(index="País", columns="Categoría", values="Beneficio", aggfunc="sum")
df_reporte_beneficios_formato_1
```

```python
#Puedo reducir la cantidad de decimales de todas las columnas a la vez, o de las columnas que especifique

df_reporte_beneficios_formato_1 = df_reporte_beneficios_formato_1.round({'Furniture': 2, 'Office Supplies': 2, 'Technology': 2})

#Igualé el dataframe a su dataframe con el redondeo para que la reducción de decimales de verdad quede guardada
```

```python
#Voy a agregar el simbolo de $ a cada Beneficio
#Y reemplazo . por , para que en el archivo excel puedan realizar cálculos normalmente

df_reporte_beneficios_formato_1["Furniture"] = df_reporte_beneficios_formato_1["Furniture"].apply(lambda x: "$"+str(x))
df_reporte_beneficios_formato_1["Furniture"] = df_reporte_beneficios_formato_1["Furniture"].apply(lambda x: x.replace(".",","))
df_reporte_beneficios_formato_1["Office Supplies"] = df_reporte_beneficios_formato_1["Office Supplies"].apply(lambda x: "$"+str(x))
df_reporte_beneficios_formato_1["Office Supplies"] = df_reporte_beneficios_formato_1["Office Supplies"].apply(lambda x: x.replace(".",","))
df_reporte_beneficios_formato_1["Technology"] = df_reporte_beneficios_formato_1["Technology"].apply(lambda x: "$"+str(x))
df_reporte_beneficios_formato_1["Technology"] = df_reporte_beneficios_formato_1["Technology"].apply(lambda x: x.replace(".",","))
```

```python
df_reporte_beneficios_formato_1
```

```python
#Exporto EL FORMATO 1 a Excel a un fichero con nombre “Reporte_Beneficios_Formato_1” y pestaña “Beneficios_cat_pais_1”

df_reporte_beneficios_formato_1.to_excel(r'C:\Mobpatec_challenge\Generados\Reporte Beneficios Formato 1.xlsx', sheet_name = "Beneficios_cat_pais")
```

```python
#AHORA REALIZARÉ UN SEGUNDO FOMATO CON GROUPBY Y DOS COLUMNAS DE AGRUPACIÓN

df_reporte_beneficios_formato_2 = pd.DataFrame(df_global_rpc.groupby(["Categoría","País"])["Beneficio"].agg("sum"))
df_reporte_beneficios_formato_2
```

```python
#Reduzco a dos decimales los elementos de Beneficio
df_reporte_beneficios_formato_2["Beneficio"] = df_reporte_beneficios_formato_2["Beneficio"].round(2)

#Voy a agregar el simbolo de $ a cada Beneficio y reemplazar . por ,
df_reporte_beneficios_formato_2["Beneficio"] = df_reporte_beneficios_formato_2["Beneficio"].apply(lambda x: "$"+str(x))
df_reporte_beneficios_formato_2["Beneficio"] = df_reporte_beneficios_formato_2["Beneficio"].apply(lambda x: x.replace(".",","))

df_reporte_beneficios_formato_2
```
